```python
''' This is a Cipher class that encodes or decodes words. Give it a try 
by passing in sentence to encode or decode and an integer to shift it by
'''


from string import ascii_lowercase

string = str(input('Please enter setnece to encode or decode: '))
shift = int(input('Please enter shift value between 1 and 25: '))
what_to_do = str(input("Please type 'encode' or 'decode': "))

class Cipher(object):
    '''
    This will encode and decode words 
    '''

    def __init__(self, string, shift):
        '''Sets the string, all_letters and shift which will be used across the functions'''
        assert type(string) == str and type(shift) == int
        self.string = string
        self.shift = shift
        self.all_letters = ascii_lowercase 
    
    
    def encode_lower_case_unicode_shifter (self, letter):  
        ''' unicode shifter function will take the letter to be shifted and the values 
        by which we shift the letter and generate a unicode to find corresponding 
        letter. 
        Function also loops back to a after loop hits z - this is for lowercase
        '''
        shifter = ord(letter) + self.shift  
        ##add in loops so unicode stays within alphabets z - 122 so this needs to loop back
        if shifter > 122: 
            shifter = (shifter - 122) + 96
            return shifter
        else: 
            return shifter 

        
    def encode_upper_case_unicode_shifter (self, letter):
        ''' unicode shifter function will take the letter to be shifted and the values 
        by which we shift the letter and generate a unicode to find corresponding 
        letter. This is for upper case   
        '''
        shifter = ord(letter) + self.shift
        if shifter > 90: 
            shifter = (shifter - 90) + 64
            return shifter
        else: 
            return shifter 
        
        
    def encode_shifted_character(self):
        '''shifted character leverages shifter function to get value associated with unicode 
        using chr function
        '''
        outcome = ''
        for letter in self.all_letters: 
            shifter_value = self.encode_lower_case_unicode_shifter(letter) 
            outcome += chr(shifter_value)

        upper_case_letters =  self.all_letters.upper()
        for letter in upper_case_letters:
            shifter_value = self.encode_upper_case_unicode_shifter(letter)
            outcome+= chr(shifter_value)

        return outcome   

    
    def encode_shifted_dictionary(self):
        '''This function will add shifted letters to corresponding letters 
        '''
        combined_letters = self.all_letters + self.all_letters.upper()

        combined_shifted_letter = self.encode_shifted_character()
        temp_dict  = {}

        ##Use zip function to loop over two lists 
        for alphabet, shifted_alpha in zip (combined_letters, combined_shifted_letter) :
            temp_dict[alphabet] = shifted_alpha

        return temp_dict


    def encode(self):
        ''' This will take the shifted dictionary from above and create an 
        ecoded string '''   
        temp_dict = self.encode_shifted_dictionary()
        string = ''

        for letter in self.string:
            if letter.isalpha() is True:
                string += temp_dict[letter]
            else: 
                string += letter
        return string
    
    
    def decode_lower_case_unicode_shifter (self, letter):
        ''' unicode shifter function will take the letter to be shifted and the values 
        by which we shift the letter and generate a unicode to find corresponding 
        letter. 
        '''
        shifter = ord(letter) - self.shift
        if shifter < 97:
            
            shifter = (shifter) + 26
            return shifter
        else: 
            return shifter 
    
    
    def decode_upper_case_unicode_shifter(self, letter):
        ''' unicode shifter function will take the letter to be shifted and the values 
        by which we shift the letter and generate a unicode to find corresponding 
        letter. 
        '''
        shifter = ord(letter) - self.shift
        print

        ##add in loops so unicode stays within alphabets z - 122 and a - 97
        if shifter < 65:
            shifter = (shifter) + 26
            return shifter
        else: 
            return shifter 
    
    
    def decode_shifted_character(self):
        '''shifted character leverages shifter function to get value associated with unicode 
        using chr function
        ''' 
        outcome = ''
        
        for letter in self.all_letters: 
            shifter_value = self.decode_lower_case_unicode_shifter(letter) 
            outcome += chr(shifter_value)
        
            upper_case_letters =  self.all_letters.upper()
        for letter in upper_case_letters:
            shifter_value = self.decode_upper_case_unicode_shifter(letter)
            outcome+= chr(shifter_value)

        return outcome   

    
    def decode_shifted_dictionary(self):
        '''This function will add shifted letters to corresponding letters 
        '''
        combined_letters = self.all_letters + self.all_letters.upper()

        combined_shifted_letter = self.decode_shifted_character()
        temp_dict  = {}

        ##Use zip function to loop over two lists 
        for letter, shifted_alpha in zip (combined_letters, combined_shifted_letter) :
            temp_dict[letter] = shifted_alpha

        return temp_dict


    def decode(self):
        ''' This will take shifted dictaionary from aboev and generate an decoded string 
        '''
        temp_dict = self.decode_shifted_dictionary()
        string = ''

        for letter in self.string:
            if letter.isalpha() is True:
                string += temp_dict[letter]
            else: 
                string += letter
        return string

##cipher creates an object  
cipher = Cipher(string, shift)
if what_to_do == 'encode':
    print(cipher.encode())
elif what_to_do == 'decode':
    print(cipher.decode())
else:
    print("Dude. Follow directions. You should've pressed 'encode' or 'decode'. Now you have to start again.")
    

```
